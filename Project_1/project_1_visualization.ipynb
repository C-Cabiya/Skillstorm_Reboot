An unexpectedly weak correlation, but VWAP does seem to be a bit more meaninful. This measures the price of a stock relevant to it's total volume; meaning that it's called the average.
```python
amin = pd.read_csv('data/am_interest.csv')
bipin = pd.read_csv('data/bip_interest.csv')
tesin = pd.read_csv('data/tesla_interest.csv')
nfin = pd.read_csv('data/netflix_interest.csv')
interest_df = {
    'amazon_interest':amin['volume'],
    'bip_interest':bipin['volume'],
    'tesla_interest':tesin['volume'],
    'netflix_interest':nfin['volume']
}
interest_df = pd.DataFrame(interest_df)
ax = sns.lineplot(data = interest_df)
ax.set(xlabel= 'Buyer Interest Over Time', ylabel = 'Volume of Transactions')
plt.xticks(visible = False)
```
